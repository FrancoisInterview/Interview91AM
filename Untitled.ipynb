```python

# 1- First step is to create dictionaries that we will use to build the translation
#Sys will be the only 3rd pary library downlaoded as STDOUT was required


import sys

OneInt={'0':'','1': 'One', '2': 'Two', '3': 'Three', '4': 'Four', '5': 'Five', \
            '6': 'Six', '7': 'Seven', '8': 'Eight', '9': 'Nine', '10': 'Ten', \
            '11': 'Eleven', '12': 'Twelve', '13': 'Thirteen', '14': 'Fourteen', \
            '15': 'Fifteen', '16': 'Sixteen', '17': 'Seventeen', '18': 'Eighteen', '19': 'Nineteen'}
TwoInt={'0':'','2': 'Twenty ', '3': 'Thirty ', '4': 'Forty ', '5': 'Fifty ', \
            '6': 'Sixty ', '7': 'Seventy ', '8': 'Eighty ', '9': 'Ninety '}
ThreeInt={'0':'','1':'One hundred ','2': 'Two hundred ', '3': 'Three hundred ', '4': 'Four hundred ', '5': 'Five hundred ', \
            '6': 'Six hundred ', '7': 'Seven hundred ', '8': 'Eight hundred ', '9': 'Nine hundred '}
Hundreds=[100,200,300,400,500,600,700,800,900]


# 2- Second step is basically how to store the number and a basic error handler to make sure nothing but integers are intered.


Input=input()
while not(Input.isdigit()):
    print('Only numbers accepted')
    print('Type any number here: ')
    Input=input()

Input=int(Input)
Input_len=len(str(Input))


# 3- Here is the definition that translate figures up to 3 numbers. 

def convert(Input):    
    if 0 <= Input <= 19:
        Input=str(Input)
        return(OneInt[Input])
    elif 20 <= Input <= 99:
        Input=str(Input)
        return(TwoInt[Input[0]])+(OneInt[Input[1]])   
    elif 100 <= Input and Input not in Hundreds :
        Input=str(Input)
        return(ThreeInt[Input[0]])+'and '+(TwoInt[Input[1]])+(OneInt[Input[2]])
    elif 100 <= Input and Input in Hundreds :
        Input=str(Input)
        return(ThreeInt[Input[0]])

# 4- Using this definition, we will integrate and maniupulate this one by slicing the number and using the definition under specific cases
# This IF statement works with numbers up to 999,999,999.
# Note: It can be hugely improved to avoid several repetitions and to go beyond the cap.


if Input_len <=3:
    Translation=convert(Input)
elif 4<=Input_len <= 6:
    #print(Input_len)
    a=Input_len-3
    Input=str(Input)
    Input2=Input[:a]
    Input2=int((Input2))
    Input=Input[-3:]
    Input=int(Input)
    Translation=convert(Input2)+" thousand "+convert(Input)
elif 7 <=Input_len <= 9:
    a=Input_len-6
    Input=str(Input)
    Input2=Input[:a]
    Input2=int((Input2))
    Input3=str((Input))
    Input=Input[-3:]
    Input=int(Input)
    Input3=Input3[3:-3]
    Input3=int(Input3)
    if Input3==0:
        Translation=convert(Input2)+" million "+convert(Input)
    else:
        Translation=convert(Input2)+" million "+convert(Input3)+" thousand "+convert(Input)


        
# 5 - Vizualising the final translation through this part        
        
stdout_fileno = sys.stdout
 
sample_input = [Translation]
 
for ip in sample_input:
    # Prints to stdout
    stdout_fileno.write(ip + '\n')
```
